```python
import pandas as pd
import matplotlib.pyplot as plt

#How we read in a pandas dataframe. The header=0 means column names are in the first row
df=pd.read_csv("netflix_titles.csv", header=0)

df.head(20)
```

#### Qual o título dos 5 primeiros filmes?

```python
#Este código dá-nos uma tabela com o título dos 5 primeiros filmes
df['title'].head(5) 
```

#### Qual o título e a duração dos 10 primeiros filmes?

```python
#Seleção de 2 colunas especificando as mesmas 
df[['title', 'duration']].head(10)
```

#### Quais os filmes lançados em anos anteriores a 1993?

```python
#Este codigo vai fazer com que sejam demonstrados todos os filmes lançados em anos anteriores a 1993
resultado = df[(df.release_year<1993)]
resultado
```

#### Quais os filmes criados entre 1974 e 1978?

```python
#Este código dá-nos os filmes lançados entre 1974 e 1978
resultado1 = df[(df.release_year>1974) & (df.release_year<1978)]
resultado1
```

#### Qual o tipo dos 15 primeiros filmes?

```python
#Este código dá-nos uma tabela com o título e o tipo de filme dos 15 primeiros filmes
df[['title', 'type']].head(15)
```

#### Gráfico de linhas do ano de lançamento dos filmes

```python
#Este código cria-nos um gráfico de linhas com os anos de lançamento dos filmes
fig, axs = plt.subplots(figsize=(15, 8))
filmes = df["release_year"].plot()
```

```python
# Gráfico de barras com a percentagem de filmes e TV Shows na Netflix
# Nas duas linhas abaixo estamos a chamar dentro da coluna tipo  os dados Movie e TV Show, estamos a usar o count para podermos determinar a quantidade relativamente a cada um 
movies = df.loc[df["type"] == "Movie"].count()[0]
tvshows = df.loc[df["type"] == "TV Show"].count()[0]

# print("Movies:", movies)
# print("TV Shows:", tvshows)

labels = ["Filmes", "TV Shows"]

explode = (.1, .1)
# Usamos o codigo plt.title para poder dar um titulo o nosso gráfico
plt.title("Movies e TV Shows na Netflix")
# Com a função pie, criar o grafico com aquela forma o nosso grafico 
plt.pie([movies, tvshows], labels = labels, autopct='%1.1f%%', explode = explode)
plt.show()
```
